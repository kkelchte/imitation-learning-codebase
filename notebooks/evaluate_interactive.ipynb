```python
import os
import numpy as np
import torch
from PIL import Image
import matplotlib.pyplot as plt
import torchvision.transforms.functional as TF
import visualpriors
import subprocess
```

```python
image_directory = '/esat/opal/kkelchte/habitat/matterport/data/v1/scans'
scene = '1pXnuDYAj8r' # 17DRP5sb8fy  1LXtFkjw3qL  1pXnuDYAj8r
image_paths = [os.path.join(image_directory, scene, scene, 'matterport_color_images', f) for f in sorted(os.listdir(os.path.join(image_directory, scene, scene, 'matterport_color_images'))) 
               if 'i1' in f]
```

```python
for image_index in range(0,len(image_paths), 10):
    image = Image.open(image_paths[image_index])
    output_images = [image.resize((256, 256), Image.NEAREST)]
    for task in ['class_object', 'class_scene', 'depth_zbuffer', 'depth_euclidean', 'segment_semantic']:
        print(f'-----{task}-----')
        o_t = TF.to_tensor(TF.resize(image, (256, 256))) * 2 - 1
        o_t = o_t.unsqueeze_(0)

        pred = visualpriors.feature_readout(o_t, task, device='cpu')
        if task == 'class_object':
            for index in torch.topk(pred[0], 5).indices:
                print(imagenet_classes[index.item()])
        elif task == 'class_scene':
            for index in torch.topk(pred[0], 5).indices:
                print(scene_classes[index.item()])
        elif task in ['depth_zbuffer', 'depth_euclidean']:
            output_images.append(TF.to_pil_image(pred[0]))
        elif task == 'segment_semantic':
            labels = semantic_masks(pred)
            output_images.append(Image.fromarray(labels))
    widths, heights = zip(*(i.size for i in output_images))

    total_width = sum(widths)
    max_height = max(heights)

    new_im = Image.new('RGB', (total_width, max_height))

    x_offset = 0
    for im in output_images:
        new_im.paste(im, (x_offset,0))
        x_offset += im.size[0]
    display(new_im)
```
